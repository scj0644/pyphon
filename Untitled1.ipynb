```python
# Python3 샘플 코드 #

import requests
import pprint
import json

url = 'http://apis.data.go.kr/1360000/AsosHourlyInfoService/getWthrDataList'
params ={'serviceKey' : 'E5m5Ih9v2H3yG9mbPWZpSWaz3a2V4JxjwQowpQSrucIJqDw4lhQ2HzL1HnxjPU8E0cnhjZSVJzpsNNITYX+VCA==', 'pageNo' : '11', 'numOfRows' : '900', 'dataType' : 'JSON', 'dataCd' : 'ASOS', 'dateCd' : 'HR', 'startDt' : '20220101', 'startHh' : '01', 'endDt' : '20230501', 'endHh' : '01', 'stnIds' : '108' }

response = requests.get(url, params=params)
print(response.content)
```

```python
import json
json_file = json.loads(response.text)
```

```python
import pandas as pd
from pandas import json_normalize
df = json_normalize(json_file['response']['body']['items']['item'])
```

```python
df
```

```python
import matplotlib.pyplot as plt

df['tm']=pd.to_datetime(df['tm'])
df['ts']=pd.to_numeric(df['ts'], downcast='float')

fig = plt.figure(figsize=(17,6))
ax = plt.subplot(1,1,1)
ax.plot(df['tm'],df['ts'],color='orange',lw=1.5,alpha=0.7)

ax.set_ylabel('Temperature', fontweight='bold', fontsize=12)
ax.set_xlabel('Date 날짜', fontweight='bold', fontsize=12)
ax.set_title('OPEN API ASOS 온도 data', fontweight = 'bold',fontsize = 14)
```

```python
import matplotlib.pyplot as plt

df['tm']=pd.to_datetime(df['tm'])
df['ta']=pd.to_numeric(df['ta'], downcast='float')

fig = plt.figure(figsize=(17,6))
ax = plt.subplot(1,1,1)
ax.plot(df['tm'],df['ta'],color='b',lw=1.5,alpha=0.7)

ax.set_ylabel('Temperature', fontweight='bold', fontsize=12)
ax.set_xlabel('Date 날짜', fontweight='bold', fontsize=12)
ax.set_title('OPEN API ASOS 온도 data', fontweight = 'bold',fontsize = 14)
```
